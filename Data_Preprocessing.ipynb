# Segment 2 - Treating Missing Values

Figure out what data is missing

```python
missing = np.nan

series_obj_2 = Series(['row 1', missing, 'row 3', 'row 4', 'row 5', 'row 6', missing, 'row 8'])
series_obj_2
```

```python
series_obj_2.isnull()
```

Filling in for missing values

```python
np.random.seed(25)

DF_obj_2 = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj_2
```

```python
#setting some value as missing
DF_obj_2.loc[3:5, 0] = missing
DF_obj_2.loc[1:4, 5] = missing
DF_obj_2
```

```python
filled_DF = DF_obj_2.fillna(0) #fill missing values with 0
filled_DF
```

```python
#using dictionary fill missing col values
#fill missing values in col 0 with 0.1, col 5 with 1.25

filled_DF = DF_obj_2.fillna({0:0.1, 5:1.25})
filled_DF
```

```python
filled_DF = DF_obj_2.fillna(method="ffill") # fill forward (last non-null element )
filled_DF
```

Counting missing values

Use: for getting summary statistics of your DF

```python
np.random.seed(25)

DF_obj_3 = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj_3.loc[3:5, 0] = missing
DF_obj_3.loc[1:4, 5] = missing
DF_obj_3
```

```python
DF_obj_3.isnull().sum()
```

Filtering out missing values

```python
DF_no_NaN =  DF_obj_3.dropna() #rows are dropped
DF_no_NaN
```

```python
DF_no_NaN =  DF_obj_3.dropna(axis=1) #cols are dropped
DF_no_NaN
```

# Segment 6 - Normalization and Standardization

```python
data = pd.read_csv("/content/auto-mpg.csv");
data.head()
```

```python
import matplotlib.pyplot as plt

mpg = data.mpg
#line chart
plt.plot(mpg)
```

```python
#Summary Statistics
mpg.describe()
```

```python
# mpg matrix -> 1 col matrix

mpg_matrix = mpg.values.reshape(-1,1)
mpg_matrix
```

```python
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale

#scale y axis ranges from 0 to 1
scaled = preprocessing.MinMaxScaler()
#scale y axis ranges from 0 to 10
# scaled = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg =  scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

Using scale() to scale your features

```python
#no mean & std dev value
standardized_mpg = scale(mpg, with_mean=False,with_std=False) # std normal distribution
plt.plot(standardized_mpg)
#got the old value back
```

# Segment 7 - Concatenating and Transforming

```python
# 6x6 matrix
# value range from 0 to 35
df_obj_7 = pd.DataFrame(np.arange(36).reshape(6,6))
df_obj_7
```

```python
df_obj_8 = pd.DataFrame(np.arange(15).reshape(5,3))
df_obj_8
```

Concatenation

```python
# join data based on row => axis=1 (concatenate by adding col)

pd.concat([df_obj_7, df_obj_8], axis=1)
```

```python
# join data based on col => (concatenate by adding row)

pd.concat([df_obj_7, df_obj_8])
```

Transformation

```python
#drop data  - by row
df_obj_7.drop([0, 2]) # removes row index 0 & 2
```

```python
#drop data  - by col
df_obj_7.drop([0, 2], axis=1) # removes col index 0 & 2
```

```python
#adding data - using join()
series_obj = Series(np.arange(6))
series_obj.name = "added_var"
series_obj
```

```python
var_added = DataFrame.join(df_obj_7, series_obj )
var_added
```

```python
#adding data - using append()

#adding the var_added var to itself.
# leave index as itself = ignore_index=False
# index reset = ignore_index=True
added_datatable = var_added.append(var_added,ignore_index=True)
added_datatable
```

```python
#sorting data - sort_values()

#by values in col 5
df_sorted = df_obj_7.sort_values(by=(5), ascending=[False])
df_sorted
```
