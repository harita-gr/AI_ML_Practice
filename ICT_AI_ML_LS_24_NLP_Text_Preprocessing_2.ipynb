```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/spam.csv", encoding='ISO-8859-1')
```

```python
data = data[["v1","v2"]]
```

Remove Punctuation

```python
import string
```

```python
punct_free = ""

def remove_punc(text):
  punc_free = ''.join([i for i in text if i not in string.punctuation])
  return punc_free
```

```python
data['clean_text'] = data['v2'].apply(lambda x:remove_punc(x))
```

```python
data.head()
```

Lowercasing

```python
data['text_lower'] = data['clean_text'].apply(lambda x:x.lower())
```

```python
data.head()
```

Tokenization

```python
import nltk
nltk.download('punkt')
```

```python
def tokenize(text):
  words = nltk.word_tokenize(text)
  return words
```

```python
data['word_token'] = data['text_lower'].apply(lambda x:tokenize(x))
```

```python
data.head()
```

Remove stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
sw = stopwords.words("english")
```

```python
def sw_removed(text):
  sw_text = [i for i in text if i not in sw]
  return sw_text
```

```python
data['no_sw'] = data['word_token'].apply(lambda x:sw_removed(x))
```

```python
data.head()
```

Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
stem_text = [];

def get_stem(words):
    # Stem each word in the list and return the new list
    stem_text = [ps.stem(word) for word in words]
    return stem_text
```

```python
data['stem_text'] = data['no_sw'].apply(lambda x:get_stem(x))
```

```python
data.head()
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemm = WordNetLemmatizer()
```

```python
lemm_text = [];

def get_lemm(words):
    # Stem each word in the list and return the new list
    lemm_text = [lemm.lemmatize(word) for word in words]
    return lemm_text
```

```python
data['lemm_text'] = data['no_sw'].apply(lambda x:get_lemm(x))
```
